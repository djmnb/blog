```python
# 将其他进制数据变成10进制
def any_to_num10(srcnum,base):
    return int(f"{srcnum}",base)
# 将十进制变成其他进制
def num10_to_any(srcnum,base):
    if base == 16:
        return hex(srcnum)
    n = 1
    num = 0
    while srcnum != 0:
        temp = srcnum % base
        num += temp*n
        srcnum //=base
        n*=10
    return num

print(any_to_num10(num10_to_any(10,16),16))
print(any_to_num10(10,2))
print()
```
